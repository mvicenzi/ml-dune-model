```python
#!/usr/bin/env python3

import h5py
import matplotlib.pyplot as plt
```

```python
fname = "../../apa-test-data/gzip2/out_monte-carlo-011984-000100_279569_164_1_20260116T164131Z/monte-carlo-011984-000100_279569_164_1_20260116T164131Z_g4-tru-anode0.h5"

with h5py.File(fname, "r") as f:
    for group in f.keys():
        print(f"Group: {group}")
        for dataset in f[group].keys():
                print(f"  Dataset: {dataset}, shape: {f[group][dataset].shape}, dtype: {f[group][dataset].dtype}")
```

```python
with h5py.File(fname, "r") as f:
    data = f["/1/frame_rebinned_reco"][:]  # shape (2560, 1500)

plt.imshow(data.T, aspect="auto", origin="lower", cmap="twilight")
plt.xlabel("Channel")
plt.ylabel("Time tick")
plt.title("frame_rebinned_reco (group 1)")
plt.colorbar(label="Value")
plt.tight_layout()
plt.show()
```

```python
with h5py.File(fname, "r") as f:
    coords = f["/1/frame_rebinned_reco/coords"][:]  
    feats = f["/1/frame_rebinned_reco/features"][:]  

    y = coords[:, 1]
    x = coords[:, 0]

    plt.scatter(x, y, c=feats, s=1, marker=".", cmap="viridis", linewidths=0)
    plt.xlim(0,2560)
    plt.ylim(0,1500)
    plt.xlabel("Channel")
    plt.ylabel("Time tick")
    plt.title("frame_rebinned_reco (group 1)")
    plt.colorbar(label="Value")
    plt.tight_layout()
    plt.show()
```
